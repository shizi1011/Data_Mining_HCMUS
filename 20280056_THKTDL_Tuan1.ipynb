## 2.1 PCA sử dụng Hàm PCA trong thư viện sklearn

Sử dụng tham số mặc định của hàm PCA : 
- n_components : số lượng components muốn giữ lại , nếu không set thì sẽ giữ lại tất cả 
- svd_solver : auto , PCA sẽ tự động dùng phương pháp nào tối ưu nhất
- ...

Một số attributes của hàm PCA :
- components_ : Principal axes trong feature space , tương ứng với hướng làm cho variance lớn nhất , sắp xếp giảm dần theo explained_variance_
- explained_variance_ : số lượng variance mà mỗi components capture được , bằng với eigenvalues của ma trận hiệp phương sai (sắp xếp giảm dần)
- explained_variance_ratio_ : giống explained_varinace_ nhưng ở dạng %
- singular_values : singular values của phép biến đổi svd trên data(đã centered)
- ...

```python
pca = PCA()
pca.fit(data2_pca)
```

Eigen Vectors ở đây là attribute components_ của PCA, và là Principal axes trong feature space, mỗi dòng của ma trận components_ tương ứng với 1 Principal Axis

```python
eigen_vectors = pca.components_
eigen_vectors
```

Eigen Values ở đây là bình phương của attribute singular_values_ của PCA, vì hàm PCA của sklearn dùng svd để tính

```python
eigen_values = np.square(pca.singular_values_)
eigen_values
```

## 2.2 PCA sử dụng phân tích SVD 

Center data về gốc tọa độ

```python
n, p = data2_pca.shape
xbar = np.array(data2_pca-np.mean(data2_pca, axis=0))
xbar
```

Phân tích SVD của ma trận Xbar 
$$ \bar{X} = U\Sigma V^T $$
Ở đây V chính là Principle Axis trong feature space = Components_ của Hàm PCA 

```python
U, S, VT = np.linalg.svd(xbar)
```

Kiểm tra so với kết quả của hàm PCA trong thư viện sklearn

Hàm numpy.allclose sẽ so sánh từng cặp giá trị (element-wise), trả về True nếu tất cả đều bằng nhau

Vì dấu của eigen vectors của PCA và dấu của eigen vector tự tính có thể không giống nhau nên em sẽ dùng numpy.abs để kiểm tra. [reference](https://math.stackexchange.com/questions/2359992/how-to-resolve-the-sign-issue-in-a-svd-problem)

```python
# Kiem tra eigen vectors
print(np.allclose(np.abs(VT.T), np.abs(eigen_vectors.T)))
# Kiem tra eigen values
print(np.allclose(np.square(S), eigen_values))
```

## 2.3 PCA sử dụng Eigen decomposition 

Dùng eigen decompositon cho ma trận hiệp phương sai 
$$ C = \frac{1}{n-1}\bar{X}^T \bar{X} = P\Lambda P^T $$

P = V trong 2.2 = components_ trong 2.1

```python
# Eigen Decomposition cho ma tran hiep phuong sai
Lambda, P = np.linalg.eig(np.cov(xbar, rowvar=False))
# Sort theo thứ tự giảm dần của eigen values
idx = Lambda.argsort()[::-1]
Lambda, P = Lambda[idx], P[:, idx]
```

Ta có : 
\begin{align}
C &= \frac{1}{n-1}\bar{X}^T \bar{X} =P\Lambda P^T\\
&= \frac{1}{n-1}(U\Sigma V^T)^T(U\Sigma V^T) \\
&= \frac{1}{n-1}V\Sigma U^T U\Sigma V^T \\
&= \frac{1}{n-1}V\Sigma^2V^T \\
&\Rightarrow \Lambda*(n-1)=\Sigma^2
\end{align}


Kiểm tra với hàm pca trong thư viện sklearn

```python

print(np.allclose(np.abs(P), np.abs(eigen_vectors.T)))
print(np.allclose(Lambda,pca.explained_variance_))
print(np.allclose((Lambda*(n-1)), eigen_values))
```
