## Tính chuẩn 1 cho từng cặp dữ liệu (theo cột)

Hàm pdist cho phép tính khoảng cách giữa các cặp điểm trong một mảng đa chiều, pdist được cài đặt hiệu quả để tính toán khoảng cách giữa các cặp điểm một cách nhanh chóng, đặc biệt đối với các mảng dữ liệu lớn

Hàm squareform sẽ chuyển đổi mảng 1D chứa khoảng cách giữa các cặp điểm (mảng được sinh ra từ hàm pdist) thành ma trận đầy đủ , trong đó giá trị tại hàng i và cột j của ma trận là giá trị khoảng cách giữa cặp điểm i và j trong mảng 1D ban đầu

```python
from scipy.spatial import distance

# Tính ma trận khoảng cách chuẩn 1 (l1 norm) bằng hàm pdist trong thư viện scipy
# Em sẽ không tự code hàm tính khoảng cách cho từng cặp điểm vì với dữ liệu lớn thì dùng hàm có sắn của thư viện (code trên ngôn ngữ C) sẽ hiệu năng tốt hơn rất nhiều 
norm1_distance = distance.squareform(distance.pdist((np.asarray(df1)).T, metric='cityblock'))

# norm1_distance[i,j] là khoảng cách chuẩn 1 (l1 norm) từ i đến j trong ma trận khoảng cách
```

Vì để tính độ đo thích ứng (match measure) và độ đo tần suất xuất hiện ngược (inverse occurrence frequency measure) dữ liệu phải là biến category nên em sẽ rời rạc hóa các thuộc tính liên tục bằng Hàm KBinsDiscretizer

```python
from sklearn.preprocessing import KBinsDiscretizer
equi_width_scaler = KBinsDiscretizer(
    n_bins=10, encode='ordinal', strategy='uniform')

df2_cp = equi_width_scaler.fit_transform(df2)
```

Để tính inverse occurence frequency measure , cần phải tính tần xuất của mỗi giá trị trong 1 cột, nên em sẽ tạo một dictionary để lưu sẵn các giá trị đó, tránh phải tính đi lại nhiều lần dẫn đến hiện năng kém

```python
value_counts_dict = {}

for i,column in enumerate(df2_cp.columns):
    value_counts_dict[i]=df2_cp[column].value_counts(normalize=True).to_dict()
value_counts_dict
```

```python
df2_cp=np.asarray(df2_cp)
```

Em sẽ dùng hàm NearestNeighbors trong thư viện sklearn để tính k láng giềng gần nhất cho mỗi điểm dữ liệu, dưới đây là mô tả đơn giản về  các tham số cũng như cách sử dụng hàm NearestNeighbor em tổng hợp từ documentation của thư viện sklearn

- n_neighbors: Đây là số lượng láng giềng gần nhất cần tìm kiếm. Tham số này được đặt bằng giá trị của k, là một số nguyên dương.

- algorithm: Đây là thuật toán được sử dụng để tìm kiếm láng giềng gần nhất. Có thể lựa chọn giữa các giá trị "auto", "ball_tree", "kd_tree", hoặc "brute". Thông thường, giá trị "auto" sẽ tự động chọn thuật toán phù hợp dựa trên dữ liệu đầu vào.

- metric: Đây là hàm khoảng cách được sử dụng để tính khoảng cách giữa các điểm dữ liệu. Có thể lựa chọn giữa các hàm khoảng cách có sẵn trong sklearn hoặc tự định nghĩa hàm khoảng cách riêng. Ví dụ: 'euclidean', 'manhattan', 'minkowski', vv.

- metric_params: Đây là tham số tùy chọn được truyền vào hàm khoảng cách nếu cần thiết. Thường thì không cần thiết phải sử dụng tham số này, và để mặc định là None.

- Sau khi khởi tạo đối tượng NearestNeighbors và fit dữ liệu vào model bằng phương thức fit(), chúng ta có thể sử dụng phương thức kneighbors() để tìm kiếm láng giềng gần nhất cho một điểm dữ liệu mới (hoặc một tập điểm dữ liệu mới). Kết quả trả về gồm hai mảng: distances là khoảng cách từ điểm dữ liệu mới đến k láng giềng gần nhất, và indices là chỉ số của các láng giềng gần nhất trong tập dữ liệu huấn luyện ban đầu.

Vì độ đo match measure và inverse occurrence frequency measure là 1 simility function, tức là sẽ cho ``giá trị càng lớn nếu 2 điểm càng gần nhau``, trong khi hàm NearestNeighbors nhận vào hàm tính khoảng cách distance function (``2 điểm càng gần nhau thì khoảng cách càng thấp``), nên lúc code hàm match_measure và inverse_occurrence_frequency_measure, em sẽ lấy giá trị âm để đúng với yêu cầu của hàm NearestNeighbors.


```python
# Hàm tính match measure giữa 2 vectors
def match_measure(X,Y):
    return -np.count_nonzero(X==Y)

def inverse_occurrence_frequency_measure(X, Y, value_counts_dict):
    # Tạo một mảng 1D chứa các giá trị frequency tương ứng với các phần tử trong X
    frequencies = np.array([value_counts_dict[i][x] for i, x in enumerate(X)])
    
    # Tìm các vị trí trong X và Y mà giá trị tại đó bằng nhau
    equal_indices = np.where(X == Y)
    
    # Tính tổng của nghịch đảo bình phương của các frequency tại các vị trí trên
    S = np.sum(1 / (frequencies[equal_indices] ** 2))
    
    return -S

```

Sử dụng hàm NearestNeighbors để tính k láng giềng gần nhất của mỗi điểm dữ liệu 

Vì dữ liệu khá lớn (~5 triệu dòng), nếu muốn tìm k láng giềng gần nhất cho mỗi điểm dữ liệu thì tương đương với việc tính 5.000.000*4.999.999/2 khoảng cách, và muốn tính mỗi khoảng cách phải loop qua tất cả các cột, tức độ phức tạp thuật toán nếu dùng cách vét cạn (brute) là O(n x n x p) hoặc O(n x log(n) x p) nếu dùng ball_tree ( với n là số dòng và p là số cột)

Với độ phức tạp thuật toán là O(n x n x p) hay O(n x log(n) x p) thì việc tính toán trên dữ liệu gần 5 triệu dòng sẽ tốn rất nhiều thời gian, nên em chỉ sẽ tính cho 1000 điểm dữ liệu đầu tiên

## Tính k láng giềng gần nhất bằng độ đo match measure 

Thuật toán em dùng để tính là ball_tree, bằng cách sử dụng ball_tree thay vì brute, việc này giảm độ phức tạp thuật toán từ O(n x n x p) xuống còn O (n x log(n) x p)


```python
from sklearn.neighbors import NearestNeighbors

neighbor_match=NearestNeighbors(n_neighbors= 20,metric=match_measure,algorithm='ball_tree').fit(df2_cp[:1000,:])
# result1_arr=neighbor_match.kneighbors(df2_cp[:1000,:])
result1_arr=neighbor_match.kneighbors()
```

Ta có result1_arr là 1 tuple gồm 2 ma trận, với ma trận đầu tiên là khoảng cách từ mỗi điểm đến k điểm gần nhất , ma trận thứ hai tương ứng là index của các điểm gần nhất đó

```python
print(result1_arr[1])
print(result1_arr[0])
```

Em sẽ kiểm tra lại bằng cách tính trực tiếp

Ở đây em chỉ kiểm tra với khoảng cách lớn nhất (37), vì ví dụ như có nhiều hơn k điểm với khoảng cách bằng nhau, thì khi lấy k điểm láng giềng có thể sẽ khác nhau

Ví dụ như có 3 điểm (1, 2, 3 ) có khoảng cách bằng nhau tới điểm 0 , thì khi lấy 2 láng giềng của điểm 0 thì có thể là (1, 2) ,(2, 3) hay (1, 3)

```python
print('k lang gieng gan nhat voi diem 0 : ')
print(np.argsort(np.array([match_measure(df2_cp[0,:],df2_cp[i]) for i in range(0,1000)]))[1:21]) # In ra k láng giềng gần nhất của điểm dữ liệu 0 (dòng 0)
print('khoang cach tuong ung')
print(-np.sort(np.array([match_measure(df2_cp[0,:],df2_cp[i,:]) for i in range(0,1000)]))[1:21]) # In ra độ đo match measure tương ứng (không lấy dấu âm)
```
